```
import pyspark.sql.functions as f
from pyspark.sql.types import StructType  
import os, json
import pandas as pd
from os.path import basename
import pyspark
import shutil, os
import delta
```

```
def get_dir_content(ls_path):
  dir_paths = dbutils.fs.ls(ls_path)
  subdir_paths = [get_dir_content(p.path) for p in dir_paths if p.isDir() and p.path != ls_path and len(p.path.split("/"))<=6]
  flat_subdir_paths = [p for subdir in subdir_paths for p in subdir]
  return list(map(lambda p: p.path, dir_paths)) + flat_subdir_paths

paths = get_dir_content('dbfs:/mnt/dlk/silver')

listPathDelaLake = []
for p in paths:
    if delta.DeltaTable.isDeltaTable(spark,p):
        jsonSchemaDict=spark.read.format("delta").load(p).schema.jsonValue()
        listPathDelaLake.append((p,json.dumps(jsonSchemaDict)))
```

```
#récupérer le modèle de la description des tables delta en json
jsonSchemaDict=spark.read.format("delta").load(listPathDelaLake[0][0]).schema.jsonValue()
jsonSchemaStr = json.dumps(jsonSchemaDict)
jsonDefSchemaDelta = f.schema_of_json(jsonSchemaStr)

df=(
    spark.createDataFrame(listPathDelaLake,['PathTable','TableJsonSchema']).select(f.col('PathTable'),f.from_json(f.col('TableJsonSchema'),jsonDefSchemaDelta).alias('TableDesc'))
    .withColumn('TableDescFields', f.explode('TableDesc.fields'))
    .select('PathTable','TableDescFields.*')
)

display(df)
df.coalesce(1).write.format("csv").mode("overwrite").option("header","true").option("delimiter",";").option("charset", "UTF-8").save("dbfs:/mnt/dlk/config-files/ADF/SchemaDeltaTables")
```
