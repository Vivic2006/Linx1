Notebook permettant de générer le fichier de schéma d'un fichier parquet.

Input Path: chemin abfss du fichier source.

Schema Path: chemin abfss du schéma généré.

Folder structure: arborescence du fichier source, par exemple: year/month/day/time.

```
%run ../Utils/COMMON
```

```
from datetime import datetime

dbutils.widgets.text('input_file_path','dbfs:/mnt/dlk/bronze/QAD/bom_mstr/','Input Path')
dbutils.widgets.text('schema_path','dbfs:/mnt/dlk/config-files/QAD_schema/bom_mstr.json','Schema Path')
dbutils.widgets.text('folder_partition', 'year=*/month=*/day=*/time=*', 'Folder structure')

folder_partition: str = dbutils.widgets.get('folder_partition')
input_full_path: str = dbutils.widgets.get('input_file_path')
schema_path: str = dbutils.widgets.get('schema_path')
```

```
df = (
  spark.read.option('basePath', input_full_path)
    .format("parquet").load(input_full_path+folder_partition+ '/*.parquet') 
)

schema: str = df.schema.json()
dbutils.fs.put(schema_path, schema, overwrite=True)
```
